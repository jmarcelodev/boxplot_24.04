```python
#Importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Carregar a planilha .csv que deesja
df1 = pd.read_csv('605_2_valid.csv')
```

```python
#Criar objeto com a produtividade e produtividade predita
produtividade = df1['produtivid']
produtividade_predita = df1['Prod_pred']
```

```python
#Plot do gráfico
plt.figure(figsize=(10, 6)) #Tamanho da figura
plt.boxplot([produtividade, produtividade_predita], labels=['Produtividade', 'Produtividade Predita']) #Obejtos presentes
plt.title(' Produtividade e Produtividade Predita') #Título
plt.ylabel('Valores'+' (t ha-¹)') #Descrição do eixo Y
plt.xlabel('Zona 605 talhão 2') #Descrição do eixo X
plt.grid(True) # Ter o grid
plt.show() 
```
